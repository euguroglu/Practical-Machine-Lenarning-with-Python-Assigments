```
import matplotlib.pyplot as plt
import math

x = [0.1*i for i in range(64)]
sin = [math.sin(i) for i in x]
cos = [math.cos(i) for i in x]

plt.plot(x,sin, label='sin(x)',color='blue', lw=2.5)
plt.plot(x,cos, label='cos(x)',color='red', lw=0.5)
plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc='upper right')
plt.show()
```

# **Modify above code to plot x^2, e^x and logx**

```
x = [0.1*i for i in range(64)]
squarex = [i**2 for i in x]

plt.plot(x,squarex, label='x^2',color='blue', lw=2.5)

plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc='upper right')
plt.show()
```

```
x = [0.1*i for i in range(64)]
expo = [math.exp(i) for i in x]

plt.plot(x,expo, label='e^x',color='red', lw=2.5)

plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc='upper right')
plt.show()
```

```
x = [0.1*i for i in range(1,64)]
  
logfunc = [math.log(i) for i in x]

plt.plot(x,logfunc, label='logx',color='red', lw=2.5)

plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc='upper right')
plt.show()
```
